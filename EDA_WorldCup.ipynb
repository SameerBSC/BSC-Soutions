```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\Dell\Desktop\worldcup.csv")
df
```

```python
df.replace("N/A", np.nan, inplace=True)
df
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.columns = df.columns.str.replace("\n"," ")
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace("  "," ")
```

```python
num_cols = [
'Previous appearances',
'Previous titles',
'Previous finals',
'Previous semifinals',
'Current FIFA rank',
'Match index',
'Match index.1',
'Match index.2',
'history with first opponent W-L',
'history with first opponent goals',
'history with second opponent W-L',
'history with second opponent goals',
'history with third opponent W-L',
'history with third opponent goals'
]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print(df)
```

```python
df['Team'] = df['Team'].replace({
'Porugal':'Portugal',
'Costarica':'Costa Rica',
'Columbia':'Colombia',
'IRAN':'Iran'
})
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))

sns.heatmap(df[num_cols].corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```
