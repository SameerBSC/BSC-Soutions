```python

class AIModel:

    def model_info(self):
        print("This is an AI Model.")



class SpamDetector(AIModel):

    def detect_spam(self):
        print("Detecting spam emails...")


class FaceRecognition(AIModel):

    def recognize_face(self):
        print("Recognizing faces...")


spam = SpamDetector()
face = FaceRecognition()

spam.model_info()
spam.detect_spam()

print()

face.model_info()
face.recognize_face()
```

```python
class SpamDetector:

    def predict(self):
        print("Predicting Spam Email")


class FaceRecognition:

    def predict(self):
        print("Recognizing Face")


class SentimentAnalyzer:

    def predict(self):
        print("Analyzing Sentiment")


models = [SpamDetector(), FaceRecognition(), SentimentAnalyzer()]

for model in models:
    model.predict()
```

```python
from abc import ABC, abstractmethod


# Abstract class
class Model(ABC):

    @abstractmethod
    def train(self):
        pass


# Child class
class SpamDetector(Model):

    def train(self):
        print("Training Spam Detector Model")


# Child class
class FaceRecognition(Model):

    def train(self):
        print("Training Face Recognition Model")


# Create objects
spam = SpamDetector()
face = FaceRecognition()

spam.train()
face.train()
```

```python
class AIModel:

    def __init__(self, name, accuracy):
        self.name = name
        self.accuracy = accuracy

    def __eq__(self, other):
        return self.accuracy == other.accuracy


model1 = AIModel("SpamDetector", 95)
model2 = AIModel("FaceRecognition", 95)

print(model1 == model2)
```

```python
# Without magic methods
class AIModel:

    def __init__(self, name, accuracy):
        self.name = name
        self.accuracy = accuracy

    def is_same_accuracy(self, other):
        return self.accuracy == other.accuracy


model1 = AIModel("SpamDetector", 95)
model2 = AIModel("FaceRecognition", 95)

print(model1.is_same_accuracy(model2))
```

```python
# Student dictionary management system
students = {}  
Books=["PythonFinalizationError","PythonBasics","PythonOOPS"]
available_books = set(Books)  

while True:
    print("\nStudent Dictionary Management System")
    print("1. Add Student")
    print("2. Borrow Book")
    print("3. Exit")

    choice = input("enter your choice :")
    if choice == "1":
        name = input("Enter student name: ")
        student_id = input("Enter student_id: ")
        students[student_id] = name
        print(f"Student {name} added successfully.")

    elif choice == "2":
        student_id = input("Enter student_id: ")
        if student_id in students:
            print(f"Available books: {', '.join(available_books)}")
            book_name = input("Enter book name to borrow: ")
            if book_name in available_books:
                available_books.remove(book_name)
                print(f"Book '{book_name}' borrowed successfully by {students[student_id]}.")
            else:
                print(f"Book '{book_name}' is not available.")
        else:
            print("Student ID not found.")
```
