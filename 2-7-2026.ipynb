```python


# Correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Remove self-correlations
corr = corr_matrix.where(~(corr_matrix == 1))

# Find the strongest positive correlation
strongest = corr.unstack().dropna().sort_values(ascending=False)

for pair, value in strongest.items():
    if pair[0] != pair[1]:
        print("Strongest Positive Correlation:")
        print(f"{pair[0]} and {pair[1]} = {value:.3f}")
        print("Interpretation: As", pair[0], "increases,", pair[1], "also tends to increase strongly.")
        break
```

```python
import pandas as pd
import numpy as np
from scipy import stats

salary = df['Salary']

# Sample statistics
mean = salary.mean()
sem = stats.sem(salary)

# 95% Confidence Interval
confidence_interval = stats.t.interval(
    confidence=0.95,
    df=len(salary)-1,
    loc=mean,
    scale=sem
)

print("Mean Salary:", mean)
print("95% Confidence Interval:", confidence_interval)
```

```python
# Independent Samples t-test (AI vs Finance Performance Rating)
import pandas as pd
from scipy.stats import ttest_ind

ai = df[df['Department'] == 'AI']['Performance_Rating']
finance = df[df['Department'] == 'Finance']['Performance_Rating']

t_stat, p_value = ttest_ind(ai, finance, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")
```

```python
# Compute Z-scores for Salary and detect outliers
import pandas as pd
from scipy.stats import zscore

df['Salary_Zscore'] = zscore(df['Salary'])

outliers = df[abs(df['Salary_Zscore']) > 2]

print(outliers)
```

```python
# Chi-Square Test (Department vs Left_Company)
import pandas as pd
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Department'], df['Left_Company'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square:", chi2)
print("P-value:", p)

if p < 0.05:
    print("Department and Left_Company are associated.")
else:
    print("No significant association.")
```

```python
# One-Way ANOVA (Monthly Sales vs Work Mode)
import pandas as pd
from scipy.stats import f_oneway

office = df[df['Work_Mode'] == 'Office']['Monthly_Sales']
hybrid = df[df['Work_Mode'] == 'Hybrid']['Monthly_Sales']
remote = df[df['Work_Mode'] == 'Remote']['Monthly_Sales']

f_stat, p_value = f_oneway(office, hybrid, remote)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Monthly Sales differ significantly across Work Modes.")
else:
    print("No significant difference.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

X = df[['Experience']]
y = df['Salary']

X_train,X_test,y_train,y_test = train_test_split(X, y,test_size=0.20,random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
mae = mean_absolute_error(y_test, y_predict)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
print("R-squared:", model.score(X, y))
print("MAE:",mae)
```
