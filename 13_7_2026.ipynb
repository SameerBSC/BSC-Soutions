```python
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()
image_name = list(uploaded.keys())[0]

image = cv2.imread(image_name)

print("Image Shape:", image.shape)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("RGB Image")
plt.axis("off")
plt.show()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
h, w = image.shape[:2]

center_x = w // 2
center_y = h // 2

cropped = image[center_y-100:center_y+100,
                center_x-100:center_x+100]

plt.imshow(cropped)
plt.title("Center Crop (200x200)")
plt.axis("off")
plt.show()
```

```python
rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
flipped = cv2.flip(image, 1)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(rotated)
plt.title("Rotated")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(flipped)
plt.title("Flipped")
plt.axis("off")

plt.show()
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

H, S, V = cv2.split(hsv)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.imshow(H, cmap='gray')
plt.title("Hue")

plt.subplot(1,3,2)
plt.imshow(S, cmap='gray')
plt.title("Saturation")

plt.subplot(1,3,3)
plt.imshow(V, cmap='gray')
plt.title("Value")

plt.show()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gaussian = cv2.GaussianBlur(image,(5,5),0)

median = cv2.medianBlur(image,5)

bilateral = cv2.bilateralFilter(image,9,75,75)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.imshow(gaussian)
plt.title("Gaussian Blur")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(median)
plt.title("Median Blur")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(bilateral)
plt.title("Bilateral Filter")
plt.axis("off")

plt.show()
```

```python
_, binary = cv2.threshold(image,127,255,cv2.THRESH_BINARY)

_, binary_inv = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(binary,cmap='gray')
plt.title("Binary")

plt.subplot(1,2,2)
plt.imshow(binary_inv,cmap='gray')
plt.title("Binary Inverse")

plt.show()
```

```python
edges = cv2.Canny(image,100,200)

plt.imshow(edges,cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary,cmap='gray')
plt.title("Binary")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

edges = cv2.Canny(gray,100,200)

plt.imshow(edges,cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

plt.imshow(sobel_x,cmap='gray')
plt.title("Sobel X")
plt.axis("off")
plt.show()

plt.imshow(sobel_y,cmap='gray')
plt.title("Sobel Y")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

laplacian = cv2.Laplacian(gray,cv2.CV_64F)

plt.imshow(laplacian,cmap='gray')
plt.title("Laplacian")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)

contours, _ = cv2.findContours(binary,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

contour_image = cv2.drawContours(image,contours,-1,(0,255,0),2)

plt.imshow(contour_image)
plt.title("Contours")
plt.axis("off")
plt.show()
```

```python
import numpy as np

kernel = np.ones((3,3),np.uint8)

eroded = cv2.erode(binary,kernel,iterations=1)

dilated = cv2.dilate(binary,kernel,iterations=1)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(eroded,cmap='gray')
plt.title("Eroded")

plt.subplot(1,2,2)
plt.imshow(dilated,cmap='gray')
plt.title("Dilated")

plt.show()
```

```python
opening  = cv2.morphologyEx(binary,cv2.MORPH_OPEN,kernel)
closing = cv2.morphologyEx(binary,cv2.MORPH_CLOSE,kernel)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(eroded,cmap='gray')
plt.title("Opening")

plt.subplot(1,2,2)
plt.imshow(dilated,cmap='gray')
plt.title("Closeing")

plt.show()
```
