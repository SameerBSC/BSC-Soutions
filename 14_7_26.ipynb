```python
import pandas as pd

data = {
    "Review": [
        "Excellent product",
        "Very bad service",
        "Amazing quality",
        "Not worth the money"
    ],
    "Sentiment": [
        "Positive",
        "Negative",
        "Positive",
        "Negative"
    ]
}

df = pd.DataFrame(data)

print(df)
```

```python
df["Sentiment"].value_counts()
```

```python
df["Review"]= df["Review"].str.lower()
print(df["Review"])
```

```python
df["Review_count"] = df["Review"].apply(len)
print(df)
```

```python
for review in df["Review"]:
 print(review)
```

```python
# df["phone_contains"] = df["review"].str.contains("Phone", case = False)
# print(df)

df["Contains_Phone"] = df["Review"].str.contains("phone", case=False)

print(df)
```

```python
positive_reviews = df[df["Sentiment"]=="positive"]
print(positive_reviews.head())

negative_reviews = df[df["Sentiment"]=="negative"]
print(negative_reviews.head())

neutral_reviews = df[df["Sentiment"]=="neutral"]
print(neutral_reviews.head())
```

```python
df["Characters"] = df["Review"].str.len()
print(df)
```

```python
df["words_column"] = df["Review"].apply( lambda x: len(x.split()))
print(df)
```

```python
import pandas as pd

data = {
    "Feedback": [
        "Excellent teaching",
        "Need more practical sessions",
        "Very supportive faculty",
        "Assignments are difficult",
        "Great classroom environment",
        "Too much homework",
        "Helpful study materials",
        "Need better lab facilities",
        "Interesting lectures",
        "Poor internet connection"
    ],
    "Sentiment": [
        "Positive",
        "Negative",
        "Positive",
        "Negative",
        "Positive",
        "Negative",
        "Positive",
        "Negative",
        "Positive",
        "Negative"
    ]
}

df = pd.DataFrame(data)

print("First Five Records:")
print(df.head())

print("\nLast Five Records:")
print(df.tail())
```

```python
df["Length"] = df["Feedback"].str.len()

longest_review = df.loc[df["Length"].idxmax()]

print(longest_review)
```
